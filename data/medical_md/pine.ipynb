```python
import os
import openai
from pinecone import Pinecone, ServerlessSpec

# Set your API keys
openai.api_key = ""
PINECONE_API_KEY= ""
pinecone_environment = 'us-east1-gcp'  # Replace with your Pinecone environment if different


# Initialize Pinecone client
pc = Pinecone(api_key=PINECONE_API_KEY, environment=pinecone_environment)
```

```python
from openai import OpenAI
client = OpenAI(api_key=openai.api_key)
def get_embedding(text, model="text-embedding-3-small"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
import glob

# Directory containing your markdown files
markdown_dir = ''

# Function to read markdown files
def read_markdown_files(directory):
    file_paths = glob.glob(os.path.join(directory, '*.md'))
    documents = []
    for file_path in file_paths:
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
            documents.append({
                'id': os.path.basename(file_path),
                'text': content
            })
            print(f'Read file: {file_path}')
            
    return documents

# Read files
documents = read_markdown_files(markdown_dir)
print(f'Read {len(documents)} documents')
```

```python

# Generate embeddings
# get_embeddings function is not defined yet
for i in documents:
    i['embedding'] = get_embedding(i['text'])

```

```python
# Check if the index already exists
index_name = 'icd11-index1'

if index_name not in pc.list_indexes():
    # Create a new index
    pc.create_index(
        spec=ServerlessSpec(cloud='aws', region='us-east-1'),
        name=index_name,
        dimension=len(documents[0]['embedding']),
        metric='cosine'  # You can choose 'euclidean' or 'dotproduct' as well
    )
```

```python
print(documents)
# 
```

```python
# Connect to the index
index = pc.Index(index_name)

    # Ensure embeddings contain the proper format (id, values, metadata)
embeddings = [
    {
        'id': doc['id'],
        'values': doc['embedding'],  # Ensure embeddings are present
        'metadata': {'filename': doc['id']}
    } 
    for doc in documents
]

# Upsert embeddings in batches
batch_size = 100
for i in range(0, len(embeddings), batch_size):
    batch = embeddings[i:i + batch_size]
    index.upsert(vectors=batch)
```

```python
# Example query to verify (embedding of a specific document)
query_embedding = embeddings[0]['values']  # Using the first document embedding as a query example

# Perform a similarity search
query_response = index.query(
    vector=query_embedding,  # The embedding of the query document
    top_k=3,                 # Retrieve the top 3 most similar embeddings
    include_metadata=True     # Include metadata for each result (like filename)
)

# Print query response with similarity results
for match in query_response['matches']:
    print(f"ID: {match['id']}, Score: {match['score']}, Metadata: {match['metadata']}")
```

```python
#        
def search_similar_documents(query_text, top_k=3):
    query_embedding = get_embedding(query_text)
    query_response = index.query(
        vector=query_embedding,
        top_k=top_k,
        include_metadata=True
    )
    return query_response
```
